```python
import pandas as pd
from rdt import HyperTransformer
from sklearn.model_selection import train_test_split

# 3️⃣ Eliminar filas con valores nulos
df_ = df.dropna()
# 4️⃣ Separar variable objetivo
target_column = "clasfinal"
X = df_.drop(columns=[target_column])
y = df_[target_column]

# 5️⃣ Crear el transformador RDT 🚀
ht = HyperTransformer()

# 6️⃣ Detectar configuración automáticamente ✅
ht.detect_initial_config(X)

# 7️⃣ Ajustar y transformar los datos 🔄
ht.fit(X)
X_transformed = ht.transform(X)

# 8️⃣ Convertir a DataFrame para visualizar 🔍
X_transformed = pd.DataFrame(X_transformed, columns=X.columns)

# 9️⃣ Mostrar comparación antes y después de la transformación
print("🔹 Datos originales:")
print(X.head())

print("\n🔹 Datos transformados:")
print(X_transformed.head())
print(X_transformed.shape)

```

```python
# prompt: selecciona las caracteristicas relevantes o con mayor varianza en X_transformed

# Calcular la varianza de cada columna en X_transformed
variances = X_transformed.var()

# Seleccionar columnas con varianza mayor que un umbral (ajusta el umbral según tu necesidad)
# Un umbral común es 0, pero puedes incrementarlo para seleccionar características con más dispersión
umbral_varianza = 0.0  # Ajusta este valor

columnas_alta_varianza = variances[variances > umbral_varianza].index.tolist()

# Opcional: Mostrar las columnas seleccionadas
print("\n🔹 Columnas con alta varianza (posibles características relevantes):")
print(columnas_alta_varianza)

# Opcional: Crear un nuevo DataFrame solo con las columnas seleccionadas
X_selected_features = X_transformed[columnas_alta_varianza]

# print("\n🔹 Datos transformados con características seleccionadas:")
# print(X_selected_features.head())
# X_selected_features.shape
```

```python
# 🔟 Separar en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)
```

```python
# clasificar con los datos transformados
from lazypredict.Supervised import LazyClassifier
from sklearn.metrics import accuracy_score

# 1️⃣1️⃣ Aplicar LazyPredict 🚀
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=accuracy_score)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)

# 1️⃣2️⃣ Mostrar resultados ordenados por precisión
print("🔹 Comparación de Modelos:")
models.sort_values(by="Accuracy", ascending=False)
```

```python

# Importar LightGBM
import lightgbm as lgb

# Crear el modelo LGBMClassifier
lgbm = lgb.LGBMClassifier(random_state=42)

# Entrenar el modelo
lgbm.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred_lgbm = lgbm.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy_lgbm = accuracy_score(y_test, y_pred_lgbm)
report_lgbm = classification_report(y_test, y_pred_lgbm)
conf_matrix_lgbm = confusion_matrix(y_test, y_pred_lgbm)

print("\n🔹 Resultados de LGBMClassifier:")
print(f"Precisión: {accuracy_lgbm:.4f}")
print("\nReporte de Clasificación:")
print(report_lgbm)
print("\nMatriz de Confusión:")
conf_matrix_lgbm

```

```python
print("\n🔹 Reporte de Métricas por Clase:")

# Extraer las métricas del reporte de clasificación
report_dict = classification_report(y_test, y_pred_lgbm, output_dict=True)

# Iterar sobre cada clase en el reporte
for class_name, metrics in report_dict.items():
    # Excluir métricas de resumen como 'accuracy', 'macro avg', 'weighted avg'
    if class_name not in ['accuracy', 'macro avg', 'weighted avg']:
        precision = metrics['precision']
        recall = metrics['recall']
        f1_score = metrics['f1-score']
        support = metrics['support']

        print(f"\nClase: {class_name}")
        print(f"  Precisión: {precision:.4f}")
        print(f"  Recall: {recall:.4f}")
        print(f"  F1-score: {f1_score:.4f}")
        print(f"  Soporte: {support}")

# Opcional: Imprimir las métricas globales (macro y weighted average)
print("\nMétricas Globales:")
print(f"  Precisión (Macro Avg): {report_dict['macro avg']['precision']:.4f}")
print(f"  Recall (Macro Avg): {report_dict['macro avg']['recall']:.4f}")
print(f"  F1-score (Macro Avg): {report_dict['macro avg']['f1-score']:.4f}")
print(f"  Precisión (Weighted Avg): {report_dict['weighted avg']['precision']:.4f}")
print(f"  Recall (Weighted Avg): {report_dict['weighted avg']['recall']:.4f}")
print(f"  F1-score (Weighted Avg): {report_dict['weighted avg']['f1-score']:.4f}")
print(f"  Precisión General: {report_dict['accuracy']:.4f}")

# Visualizar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_lgbm, annot=True, fmt='d', cmap='Blues', xticklabels=lgbm.classes_, yticklabels=lgbm.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Matriz de Confusión para LGBMClassifier')
plt.show()

```
